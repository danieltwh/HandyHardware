```python
conn.execute("DROP SCHEMA IF EXISTS ECOMMERCE;")
conn.execute("CREATE DATABASE IF NOT EXISTS ECOMMERCE;")
conn.execute("USE ECOMMERCE")

conn.execute("""
CREATE TABLE Customers(customerID VARCHAR(100) NOT NULL,
					customerName VARCHAR(100) NOT NULL,
					gender ENUM('MALE', 'FEMALE', 'OTHERS') NOT NULL,
					email VARCHAR(100) NOT NULL,
					phoneNumber VARCHAR(8) NOT NULL,
					address VARCHAR(100) NOT NULL,
					customerPassword VARCHAR(30) NOT NULL,
					PRIMARY KEY (customerID));
                    

CREATE TABLE Products(category VARCHAR(6) NOT NULL,
					cost DECIMAL(10,2) NOT NULL,
					model VARCHAR(10) NOT NULL,
					price DECIMAL(10,2) NOT NULL,
                    productID INTEGER NOT NULL,
					warrantyMonths INTEGER NOT NULL,
                    PRIMARY KEY (productID));
                    

CREATE TABLE Administrators(administratorID VARCHAR(100) NOT NULL,
					administratorName VARCHAR(100) NOT NULL,
					gender ENUM('MALE', 'FEMALE', 'OTHERS') NOT NULL,
					phoneNumber VARCHAR(8) NOT NULL,
					administratorPassword VARCHAR(30) NOT NULL,
                    PRIMARY KEY (administratorID));
                    
CREATE TABLE Items(itemID INTEGER NOT NULL,
                    colour VARCHAR(10) NOT NULL,
                    factory VARCHAR(100) NOT NULL,
                    powerSupply VARCHAR(100) NOT NULL,
                    purchaseStatus ENUM('SOLD', 'UNSOLD') NOT NULL,
                    productionYear VARCHAR(100) NOT NULL,
					productID INTEGER NOT NULL,
                    PRIMARY KEY (itemID),
                    FOREIGN KEY (productID) REFERENCES Products(productID));

CREATE TABLE Payments(paymentID INTEGER NOT NULL,
					itemID INTEGER NOT NULL,
					purchaseDate DATE NOT NULL,
					customerID VARCHAR(100) NOT NULL,
                    amount DECIMAL(10,2) NOT NULL,
                    PRIMARY KEY (paymentID),
                    FOREIGN KEY (customerID) REFERENCES Customers(customerID),
                    FOREIGN KEY (itemID) REFERENCES Items(itemID));
                    
CREATE TABLE Requests(requestID INTEGER NOT NULL,
					itemID INTEGER NOT NULL,
					administratorID VARCHAR(100) NOT NULL,
					requestStatus ENUM('Submitted', 'Submitted and Waiting for payment', 'In progress', 'Approved', 'Canceled', 'Completed') NOT NULL,
                    requestDetails VARCHAR(200) NOT NULL,
                    PRIMARY KEY (requestID),
                    FOREIGN KEY (itemID) REFERENCES Items(itemID),
                    FOREIGN KEY (administratorID) REFERENCES Administrators(administratorID));
                    
CREATE TABLE ServiceFees(requestID INTEGER NOT NULL,
                    amount DECIMAL(10,2) NOT NULL,
					creationDate DATE NOT NULL,
					settlementDate DATE NULL,
                    FOREIGN KEY (requestID) REFERENCES Requests(requestID),
                    PRIMARY KEY (requestID, creationDate));
                    

CREATE TABLE Services(requestID INTEGER NOT NULL,
					serviceStatus ENUM('Waiting for approval', 'In progress', 'Completed') NOT NULL,
                    FOREIGN KEY (requestID) REFERENCES Requests(requestID),
                    PRIMARY KEY (requestID, serviceStatus));
""")
```
